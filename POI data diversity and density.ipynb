### Compute POI diversity

```python
df_poi_category=df_poi.groupby('DataZone')['categoryname'].value_counts()
df_poi_category.head(50)
```

```python

df_poi_category = pd.DataFrame(df_poi_category)
df_poi_category.rename(columns = {'categoryname':'count'}, inplace = True)
df_poi_category.head(20)
```

```python
df_poi_category_1 = df_poi_category.reset_index("categoryname")
```

```python
df_poi_category_1
```

```python
df_poi_category_1.index
dz_list = list(df_poi_category_1.index)
dz_list = list(dict.fromkeys(dz_list))
dz_list
```

```python
test = list(df_poi_category.loc['S01009775']['count'])
test
```

#### Shannon entropy

```python
#shannon entropy
import math
poi_shannon = []
for dz in dz_list:
    shannon_entropy_value = 0
    summa = df_poi_category.loc[dz]['count'].sum()
    count_list =  list(df_poi_category.loc[dz]['count'])
    for i in count_list:
        p_i = i/summa
        entropy_i = p_i * (math.log(p_i, 2))
        shannon_entropy_value += entropy_i
    shannon_entropy_value = round(shannon_entropy_value,2)
    poi_shannon.append(-shannon_entropy_value)
print(poi_shannon)
#https://onestopdataanalysis.com/shannon-entropy/
    
```

#### Simpson index

```python
#simpson index
poi_simpson = []
for dz in dz_list:
    sum_simpson_i = 0
    summa = df_poi_category.loc[dz]['count'].sum()
    count_list =  list(df_poi_category.loc[dz]['count'])
    for i in count_list:
        s_i = i/summa
        simpson_i = s_i**2
        sum_simpson_i += simpson_i
        simpson_index = 1-sum_simpson_i
        simpson_index = round(sum_simpson_i,2)
    poi_simpson.append(simpson_index)
print(poi_simpson)
```

#### Richness

```python
series_category = df.groupby('DataZone')['categoryname'].unique()

df_category = list(series_category)

category_count = []
for i in df_category:
    category_c = len(i)
    category_count.append(category_c)

```

```python
dic_final = {'DataZone':dz_list,'Richness':category_count,'Shannon entropy':poi_shannon,'Simpson index':poi_simpson}
df_final = pd.DataFrame(dic_final)
df_final
```

```python
df_final = pd.merge(df_final, df2, on='DataZone')
df_final
```
